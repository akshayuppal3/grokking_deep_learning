```python
## tring basic neural network code from grokking to update weights
weight = 0.5
input = 0.5
goal = 0.8

step_amount = 0.001

def get_pred(input, weight):
    return input * weight

for i in range(2201):  # 1101 is when it terminates
    pred = weight * input
    error = (pred - goal)**2
    
    print(f'error = {error} prediction= {pred} ')
    up_error = (get_pred(input,weight + step_amount) - goal)**2
    down_error = (get_pred(input,weight - step_amount) - goal)**2
    
    if up_error < down_error:
        weight += step_amount
        print(f'{weight} updated up {pred}',)
    else:
        print(f'{weight} updated down {pred}',)
        weight -= step_amount
print(pred)
```

```python
weight = 0.5
input = 2
goal = 0.8
alpha = 0.1

for i in range(8):
    pred = weight * input
    error = (pred - goal)**2
    delta = pred - goal
    weight_delta = delta * input
    weight -= alpha * weight_delta
    
    print(f'pred = {pred} error = {error}  weight = {weight}')
```
